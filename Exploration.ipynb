```python
def print_stats(df):
    # some stats per column
    stats = {
        'col': [],
        'missing_ratio': [],
        'unique_cnt': [],
        'dtype': [],
        'max': [],
        'min': []
    }
    for col in df.columns:
        stats['col'].append(col)
        missing_ratio = df[col].isnull().sum()/len(df)
        stats['missing_ratio'].append(missing_ratio)
        stats['unique_cnt'].append(len(df[col].unique()))
        stats['dtype'].append(df[col].dtype)
        if pd.api.types.is_numeric_dtype(df[col].dtype):
            stats['max'].append(df[col].max())
            stats['min'].append(df[col].min())
        else:
            stats['max'].append('na')
            stats['min'].append('na')
    return pd.DataFrame(stats)
```

```python
class TrainSet(torch.utils.data.Dataset):
    
    def __init__(self, df, transform=None):
        self.df_x = df.drop(columns=['Ownership_Indicator', 'id', 'set']).copy()
        self.df_y = df['Ownership_Indicator'].copy()
        self.ids = df['id'].copy()
        self.transform = transform
    
    def __len__(self):
        return len(self.df_x)

    def __getitem__(self, idx):
        x = torch.tensor(self.df_x.iloc[idx].values, dtype=torch.float)
        y = torch.tensor(self.df_y.iloc[idx], dtype=torch.float)
        return x, y
```

```python
# train_set = TrainSet(df_train_sm_train)
# val_set = TrainSet(df_train_sm_val)
train_set = TrainSet(df_train_train)
val_set = TrainSet(df_train_val)
sample = train_set[0]
x, y = sample
print(x.shape)
print(y.shape)
print(y)
```

```python
train_loader = torch.utils.data.DataLoader(train_set, batch_size=128, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_set, batch_size=128, shuffle=False)
```

```python
y_cnt = 0
for x, y in train_loader:
    y_cnt += y.sum()
    print(y_cnt)
print(y_cnt)
```

```python
class Model(torch.nn.Module):
    
    def __init__(self):
        super(Model, self).__init__()
        self.relu = torch.nn.ReLU()
        self.lin1 = torch.nn.Linear(in_features=25, out_features=64)
        self.bn1 = torch.nn.BatchNorm1d(64)
        self.lin2 = torch.nn.Linear(in_features=64, out_features=64)
        self.bn2 = torch.nn.BatchNorm1d(64)
        self.lin3 = torch.nn.Linear(in_features=64, out_features=10)
        self.bn3 = torch.nn.BatchNorm1d(10)
        self.lin4 = torch.nn.Linear(in_features=10, out_features=1)
    
    def forward(self, x):
        x = self.relu(self.bn1(self.lin1(x)))
        x = self.relu(self.bn2(self.lin2(x)))
        x = self.relu(self.bn3(self.lin3(x)))
        x = self.lin4(x)
        return x
```

```python
device = torch.device('cuda')
model = Model().to(device)
criterion = torch.nn.BCEWithLogitsLoss()
optim = torch.optim.Adam(model.parameters(),lr=0.01)

for epoch in range(10):
    model.train()
    train_loss = 0.0
    for i, batch in enumerate(train_loader):
        x, y = batch
        y_hat = model(x.float().to(device))
        optim.zero_grad()
        loss = criterion(y_hat, y.view(-1, 1).to(device))
        loss.backward()
        optim.step()
        train_loss += loss.item()
    
    model.eval()
    val_loss = 0.0
    acc_avg = 0.0
    for i, batch in enumerate(val_loader):
        x, y = batch
        y_hat = model(x.float().to(device))
        loss = criterion(y_hat, y.view(-1, 1).to(device))
        
        y_hat = torch.sigmoid(y_hat).view(-1)
#         print("printing y =================")
#         print(y)
#         print("printing y_hat before round =================")
#         print(y_hat)
        y_hat = torch.round(y_hat)
#         print("printing y_hat before round =================")
#         print(y_hat)

        correct = (y.to(device) == y_hat).sum().float()
#         print(f"correct {correct}")
        acc = correct/y.shape[0]
        acc_avg += acc
        val_loss += loss.item()
    acc_avg = torch.round(acc_avg * 100 / len(train_loader))
    print(f"Epoch: {epoch}")
    print(f"Acc: {acc_avg}")
    print(f"loss train: {train_loss}")
    print(f"loss val: {val_loss}")
```
