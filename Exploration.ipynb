```python
def print_stats(df):
    # some stats per column
    stats = {
        'col': [],
        'missing_ratio': [],
        'unique_cnt': [],
        'dtype': [],
        'max': [],
        'min': []
    }
    cols_to_remove = [] # Remove columns with 80% missing value
    for col in df.columns:
        stats['col'].append(col)
        missing_ratio = df[col].isnull().sum()/len(df)
        if missing_ratio > 0.75:
            cols_to_remove.append(col)
        stats['missing_ratio'].append(missing_ratio)
        stats['unique_cnt'].append(len(df[col].unique()))
        stats['dtype'].append(df[col].dtype)
        if pd.api.types.is_numeric_dtype(df[col].dtype):
            stats['max'].append(df[col].max())
            stats['min'].append(df[col].min())
        else:
            stats['max'].append('na')
            stats['min'].append('na')
    return pd.DataFrame(stats)
```

```python
class TrainSet(torch.utils.data.Dataset):
    
    def __init__(self, df, transform=None):
        self.df_x = df.drop(columns=['Ownership_Indicator', 'id', 'set'])
        self.df_y = df['Ownership_Indicator']
        self.ids = df['id']
        self.transform = transform
    
    def __len__(self):
        return len(self.df_x)

    def __getitem__(self, idx):
        x = torch.tensor(self.df_x.iloc[idx].values, dtype=torch.float)
        y = torch.tensor(self.df_y.iloc[idx], dtype=torch.float)
        return x, y
```

```python
train_set = TrainSet(df_train)
sample = train_set[0]
x, y = sample
print(x.shape)
print(y.shape)
print(y)
```

```python
train_loader = torch.utils.data.DataLoader(train_set, batch_size=100000, shuffle=True)
```

```python
y_cnt = 0
for x, y in train_loader:
    y_cnt += y.sum()
    print(y_cnt)
print(y_cnt)
```

```python
class Model(torch.nn.Module):
    
    def __init__(self):
        super(Model, self).__init__()
        self.relu = torch.nn.ReLU()
        self.lin1 = torch.nn.Linear(in_features=25, out_features=10)
        self.bn1 = torch.nn.BatchNorm1d(10)
        self.lin2 = torch.nn.Linear(in_features=10, out_features=1)
    
    def forward(self, x):
        x = self.relu(self.bn1(self.lin1(x)))
        x = self.lin2(x)
        return x
```

```python
model = Model()
criterion = torch.nn.BCEWithLogitsLoss()
optim = torch.optim.Adam(model.parameters())

for epoch in range(1):
    model.train()
    for i, batch in enumerate(train_loader):
        x, y = batch
        print(x.shape, y.shape)
        y_hat = model(x.float())
        print(y_hat.shape)
        optim.zero_grad()
        loss = criterion(y_hat, y.view(-1, 1))
        loss.backward()
        optim.step()
        print(f"{i}: {loss.item()}")
```
